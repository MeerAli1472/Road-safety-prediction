```python
from tkinter import*
from tkinter import ttk
from PIL import Image,ImageTk
from tkinter import messagebox
from predictive_model import Predict
import seaborn as sns
```

```python
class login:
    def __init__(self,root):
        self.root=root
        self.root.geometry("1560x768+0+0")
        self.root.minsize(1024,768)
        self.root.title("Login Page")
        self.root.configure(bg="cyan4")
        
        self.var_username = StringVar()
        self.var_password = StringVar()
        
        label1 = Label(self.root,text="Login Page",bg="cyan4",fg="cyan",font=("Helvetica",30,"bold"))
        label1.place(x=600,y=20)

        frame=LabelFrame(self.root,bd=4,bg="cyan3",relief=RIDGE,text="Login",font=("times new roman",12,"bold"))
        frame.place(x=450,y=100,width=500,height=550)
        
        
        #Username label
        username_label=Label(frame,text="Username :",font=("times new roman",12,"bold"),bg="cyan4",fg="white")
        username_label.place(x=100,y=150)
        
        #Username entrybox
        Username_entry = ttk.Entry(frame,width=15,textvariable=self.var_username,font=("times new roman",12,"bold"))
        Username_entry.place(x=200,y=150)
        
        #Password label
        Password_label=Label(frame,text="Password  :",font=("times new roman",12,"bold"),bg="cyan4",fg="white")
        Password_label.place(x=100,y=200)
        
        #Username entrybox
        Password_entry = ttk.Entry(frame,width=15,textvariable=self.var_password,font=("times new roman",12,"bold"),show='*')
        Password_entry.place(x=200,y=200)
        
        
        btn = Button(frame,text="Login",command = self.log,font=("times new roman",13,"bold"),bg="green",fg="white")
        btn.place(x=200,y=250)
        
    def log(self):
        username = self.var_username.get()
        password = self.var_password.get()
        user = "admin"
        passw = "12345"

        if username == user and password == "12345":
            messagebox.showinfo("Success", "Successfully Login", parent=root)
            self.root.withdraw()  # Hide the login window
            self.predict()
        else:
            messagebox.showerror("Error", "Please provide correct info", parent=root)
        
            
            
    def predict(self):
        self.new_window = Toplevel(self.root)
        self.app = Predict(self.new_window)
        self.new_window.protocol("WM_DELETE_WINDOW", self.exit_application)

    def exit_application(self):
        self.root.destroy()

        
if __name__ == "__main__":                                 
    root=Tk()
    obj = login(root)
    root.mainloop()
```
